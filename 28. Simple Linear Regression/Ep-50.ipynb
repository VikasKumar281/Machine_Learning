```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('placement2.csv')
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package( in lpa )')
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train,y_test = train_test_split(X,y , test_size = 0.2 , random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(X_train , y_train)
```

```python
X_test
```

```python
y_test
```

```python
# y_pred = lr.predict(X_test.iloc[0].values.reshape(1,1))
# y_pred = lr.predict(X_test.iloc[1].values.reshape(1,1))
y_pred = lr.predict(X_test.iloc[2].values.reshape(1,1))
```

```python
y_pred
```

```python
plt.scatter(df['cgpa'],df['package'])


# It will plot the Best Fit line -------------------------------------------------------------------------------------->

# plt.plot(X_test,lr.predict(X_test), color='red')
plt.plot(X_train,lr.predict(X_train), color='red')

plt.xlabel('CGPA')
plt.ylabel('Package( in lpa )')
```
