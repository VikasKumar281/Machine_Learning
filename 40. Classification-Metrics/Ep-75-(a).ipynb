# Classification Metrics(Part-1) =>

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("heart.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(df.iloc[:,0:-1] , df.iloc[:,-1], test_size = 0.2 , random_state = 2)
```

### df.iloc[:, 0:-1] =>	Selects all rows and all columns except the last column. This is your features (X).
### df.iloc[:, -1] =>	Selects all rows and only the last column. This is your target (y).

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
clf1.fit(X_train , y_train)
clf2.fit(X_train , y_train)
```

```python
y_pred1 = clf1.predict(X_test)
y_pred2 = clf2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix

print("Accuracy of Logistic Regression: " , accuracy_score(y_test , y_pred1))
print("Accuracy of Decision Tree Classifier: " , accuracy_score(y_test , y_pred2))
```

## Confusion Matrix=>

```python
confusion_matrix(y_test , y_pred1)
```

```python
confusion_matrix(y_test , y_pred2)
```

# Confusion Matrix for Multi-Classification Problem =>

```python
df2 = pd.read_csv("iris.csv")
```

```python
df2.sample(5)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df2['species'] = encoder.fit_transform(df2['species'])
```

```python
df2.sample(5)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df2.iloc[:,0:-1],df2.iloc[:,-1],test_size=0.2,random_state=1)
```

```python
clf3 = LogisticRegression()
clf4 = DecisionTreeClassifier()
```

```python
clf3.fit(X_train , y_train)
clf4.fit(X_train , y_train)
```

```python
print(y_test.shape)
y_pred3 = clf3.predict(X_test)
y_pred4 = clf4.predict(X_test)
print(y_pred3)
print(y_pred3.shape)


print(y_pred4)
```

```python
print("Accuracy of Logistic Regression: " , accuracy_score(y_test , y_pred3))
print("Accuracy of Decision Tree Classifier: " , accuracy_score(y_test , y_pred4))
```

```python
print("Logistic Regression Confusion Matrix:\n ")
pd.DataFrame(confusion_matrix(y_test , y_pred3) , columns = list(range(0,3)))
```

```python
print("Decision Tree Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred4),columns=list(range(0,3)))
```
