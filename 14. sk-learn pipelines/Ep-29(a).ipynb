# Project  Using Pipelines--------------------->

```python
df2 = pd.read_csv('train.csv') 
```

```python
df2.drop(columns =['PassengerId','Name','Ticket','Cabin'],inplace = True)
```

```python
df2.head()
```

```python
# Step-1 => train/test/split

X_train,X_test,y_train,y_test = train_test_split(df2.drop(columns =['Survived']),df2['Survived'],test_size = 0.2,random_state = 42)
# random_state ka use hm data ko har baar ek jaisa baatne ke liye karte hai
```

```python
X_train.head()
```

```python
X_train.isnull().sum()
```

```python
# Imputation Transformer---------------------------->
trf1 = ColumnTransformer([
    ('impute_age',SimpleImputer(),[2]),
    ('impute_embarked',SimpleImputer(strategy = 'most_frequent'),[6])
],remainder ='passthrough')

# Note => Here [2] and [6] are the Index numbers of 'age' and 'embarked' 
```

```python
# One Hot Encoding------------------------------------->\
trf2 = ColumnTransformer([
    ('ohe_sex_embarked',OneHotEncoder(sparse_output = False,handle_unknown ='ignore'),[1,6])
],remainder = 'passthrough')
```

```python

# Scaling -------------------->
trf3 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])
```

```python
# Feature Selection ------------------------->
trf4 = SelectKBest(score_func = chi2,k=8)

# Here k=8 means hm top 8 Columns ka use kar rhe hai out of 10 columns
```

```python
# Train the Model--------------------------------------->
trf5 = DecisionTreeClassifier()
```

## Create Pipeline

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5)
])
```

```python
# Pipeline ki jagah ham make_pipeline bhi use kar sakte kyuki yeh simple hota hai----------------------->
# Same applies to ColumnTransformer() vs make_column_transformer()---------------------------->

#Alternate syntax
# pipe = make_pipeline(trf1,trf2,trf3,trf4,trf5)
```

```python
# train--------------->

# Agar aap Pipeline ke through model train nhi kar rhe ho to aap fit_transform ka use karte ho but here we train the model using Pipelin so we used 'fit'

pipe.fit(X_train,y_train)
```
